```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

print(f"No of schools: {len(school_data)}")
print(f"No of schools: {len(student_data)}")
school_data
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Part 1: Local Government Area (LGA) Summary
# number of schools
iNoOfSchools = len(school_data)

#total students
iTotalStudents = len(student_data)

#total budget of schools
iTotalBudget = school_data["budget"].sum()

# avg maths score for all students
fAvgMathScore = float(student_data["maths_score"].mean())

# avg reading score for all students
fAvgRdgScore = float(student_data["reading_score"].mean())

# Get no of students with maths score > 50 and calculate % of students passing maths
MathsPercent = student_data.loc[(student_data["maths_score"] >= 50)]
MathsPercent

fPercentMaths = float((len(MathsPercent) *100 / iTotalStudents))

# Get no of students with reading score > 50 and calculate % of students passing maths
RdgPercent = student_data.loc[(student_data["reading_score"] >= 50)]
RdgPercent

fPercentRdg = float((len(RdgPercent) *100 / iTotalStudents))
                    
# Get no of students with reading AND maths score > 50 and calculate % of students passing both
OverallPercent = student_data.loc[((student_data["reading_score"] >= 50) &
                                       (student_data["maths_score"] >= 50))]
OverallPercent

fPercentOverall = float((len(OverallPercent) *100 / iTotalStudents))                

df_AreaSummary = pd.DataFrame([
    {"Total Schools": iNoOfSchools,
     "Total Students": iTotalStudents,
     "Total budget ($)": iTotalBudget,
     "Average maths score": fAvgMathScore,
     "Average reading score": fAvgRdgScore,
     "% passing maths": fPercentMaths,
     "% passing reading": fPercentRdg,
     "% overall passing": fPercentOverall}
])

df_AreaSummary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:

```python
# Part 2: School Summary
school_data_complete
```

```python
# Find avg maths score for each school
avg_math_score = school_data_complete.groupby('school_name')['maths_score'].mean()
avg_math_score 
```

```python
# Find avg reading score for each school
avg_rdg_score = school_data_complete.groupby('school_name')['reading_score'].mean()
avg_rdg_score 
```

```python
# Find number of students passing maths for each school
maths_pct_df = pd.DataFrame(student_data.loc[(student_data["maths_score"] >= 50)])
maths_pct_df = pd.DataFrame(maths_pct_df['school_name'].value_counts())

# reset index to column to merge data
maths_pct_df.reset_index(inplace=True)
maths_pct_df = maths_pct_df.rename(columns = {'index':'school_name'})
maths_pct_df
```

```python
# Find number of students passing reading for each school
rdg_pct_df = pd.DataFrame(student_data.loc[(student_data["reading_score"] >= 50)])
rdg_pct_df = pd.DataFrame(rdg_pct_df['school_name'].value_counts())

rdg_pct_df.reset_index(inplace=True)
rdg_pct_df = rdg_pct_df.rename(columns = {'index':'school_name'})
rdg_pct_df
```

```python
# Find number of students passing reading AND maths for each school
overall_pct_df = pd.DataFrame(student_data.loc[((student_data["reading_score"] >= 50) &
                                              (student_data["maths_score"] >= 50))])
overall_pct_df = pd.DataFrame(overall_pct_df['school_name'].value_counts())

overall_pct_df.reset_index(inplace=True)
overall_pct_df = overall_pct_df.rename(columns = {'index':'school_name'})
overall_pct_df
```

```python
# Calculate pct of students passing maths, reading and overall
combine_passing_df = pd.merge(maths_pct_df, rdg_pct_df, on=["school_name", "school_name"])
combine_passing_df = pd.merge(combine_passing_df, overall_pct_df, on=["school_name", "school_name"])

combine_passing_df =combine_passing_df.rename(columns={"count_x": "No maths", "count_y": "No reading", "count": "No overall"})
combine_passing_df
```

```python
# Create dataframe from groupby object
AvgScore_df = pd.DataFrame(
    {"Average maths score": avg_math_score,
     "Average reading score": avg_rdg_score}
)

AvgScore_df.reset_index(inplace=True)
AvgScore_df = AvgScore_df.rename(columns = {'index':'school_name'})
AvgScore_df
```

```python
# list of student budget for each school = total school budget/ total students
lStudentBudget = (school_data['budget']/school_data['size']).to_list()
#lStudentBudget

# create school summary dataframe and add new column
school_summary_df = school_data.rename(columns={"type":"School Type", "size":"Total Students", "budget":"Total School Budget"})
school_summary_df["Per Student Budget"] = lStudentBudget
school_summary_df
```

```python
school_summary_df = pd.merge(school_summary_df, AvgScore_df, on=["school_name", "school_name"])
school_summary_df
```

```python
school_summary_df = pd.merge(school_summary_df, combine_passing_df, on=["school_name", "school_name"])
school_summary_df
```

* Sort and display the top five performing schools by % overall passing.

```python
# calculate pct of passing for maths, reading and overall for each school
school_summary_df["% Passing Maths"] = school_summary_df["No maths"] * 100 / school_summary_df["Total Students"]
school_summary_df["% Passing Reading"] = school_summary_df["No reading"] * 100 / school_summary_df["Total Students"]
school_summary_df["% Passing Overall"] = school_summary_df["No overall"] * 100 / school_summary_df["Total Students"]

# drop not required columns and set index to school name and add currency formatting to budget columns
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df = school_summary_df.drop(columns=["School ID", "No maths", "No reading", "No overall"])
school_summary_df = school_summary_df.rename(columns={"school_name":"School Name"})
school_summary_df = school_summary_df.set_index(['School Name'])
school_summary_df
```

```python
# Top 5 performing schools
top_5_best_df = school_summary_df.sort_values("% Passing Overall", ascending=False)
top_5_best_df.head()
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Find number of students passing reading AND maths for each school
# Filter year 9
year_9_df = pd.DataFrame(student_data.loc[student_data["year"] == 9])
year_9_df

# group by school name and find mean
math_year_9 = year_9_df.groupby('school_name')['maths_score'].mean()

# Filter year 10
year_10_df = pd.DataFrame(student_data.loc[student_data["year"] == 10])
year_10_df

# group by school name and find mean
math_year_10 = year_10_df.groupby('school_name')['maths_score'].mean()

# Filter year 11
year_11_df = pd.DataFrame(student_data.loc[student_data["year"] == 11])
year_11_df

# group by school name and find mean
math_year_11 = year_11_df.groupby('school_name')['maths_score'].mean()

# Filter year 12
year_12_df = pd.DataFrame(student_data.loc[student_data["year"] == 12])
year_12_df

# group by school name and find mean
math_year_12 = year_12_df.groupby('school_name')['maths_score'].mean()

# Create dataframe from groupby objects for maths year wise scores
AvgMathsScore_df = pd.DataFrame(
    {"Year 9": math_year_9,
     "Year 10": math_year_10,
     "Year 11": math_year_11,
     "Year 12": math_year_12}
)

AvgMathsScore_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Summary by school spending 
# Create the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# convert per student budget from string to float
school_summary_df["Student Budget"] = school_summary_df["Per Student Budget"].str.replace('$', '')
school_summary_df["Student Budget"] = school_summary_df["Student Budget"].astype(float)

# Slice the data and place it into bins
score_spending_df = school_summary_df
score_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Student Budget"], spending_bins, labels=labels, include_lowest=True)

# Creating a group based off of the bins
spending_score = score_spending_df.groupby("Spending Ranges (Per Student)")
print(spending_score["Student Budget"].count())

# Get the average of each of the first 5 rating columns within the GroupBy object
spending_score[["Average maths score", "Average reading score", "% Passing Maths", "% Passing Reading", "% Passing Overall"]].mean()

```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Summary by school spending 
# Create the bins 
bins = [0, 999, 1999, 4999]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Slice the data and place it into bins
school_size_df = school_summary_df
school_size_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins, labels=labels, include_lowest=True)

# Creating a group based off of the bins
schoolsize_score = school_size_df.groupby("School Size")
print(schoolsize_score["Total Students"].count())

# Get the average of each of the first 5 rating columns within the GroupBy object
schoolsize_score[["Average maths score", "Average reading score", "% Passing Maths", "% Passing Reading", "% Passing Overall"]].mean()

```
